```python
# 4번 과제

from skimage.io import imread
import scipy.fftpack as fp
from scipy import ndimage
import matplotlib.pyplot as plt
import numpy as np
import numpy.fft as fp
from matplotlib import cm
```

```python
plt.gray()
im = np.mean(imread('./lena.jpg'), axis=2)
freq = fp.fft2(im)
freq_gaussian = ndimage.fourier_gaussian(freq, sigma=4)
im1 = fp.ifft2(freq_gaussian)
im2 = fp.fft2(im1)
plt.figure(figsize=(12, 12))
plt.subplot(2, 3, 1); plt.imshow(im)
plt.subplot(2, 3, 2); plt.imshow((20*np.log10( 0.1 + fp.fftshift(freq))).astype(int))
plt.subplot(2, 3, 3); plt.imshow((20*np.log10( 0.1 + fp.fftshift(freq_gaussian))).astype(int))
plt.subplot(2, 3, 4); plt.imshow(im1.real)
plt.subplot(2, 3, 5); plt.imshow((20*np.log10( 0.1 + fp.fftshift(im2))).astype(int))
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.gca(projection='3d')
Y = np.arange(im.shape[0])
X = np.arange(im.shape[1])
X, Y = np.meshgrid(X, Y)
Z = (20*np.log10( 0.1 + fp.fftshift(freq))).astype(int)
ax.plot_surface(X,Y,Z, cmap=cm.coolwarm)
```
