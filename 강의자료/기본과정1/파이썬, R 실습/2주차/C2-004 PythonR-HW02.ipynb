#**Quiz 1**
Kaka* T와 같은 택시 매칭 어플 내 새로운 기능을 개발하고자 합니다.

50명의 승객과 매칭 기회가 있었을 때, 총 탑승 승객 수를 구하는 프로그램을 작성하세요.

* 조건1: 승객별 운행 소요 시간은 5분 ~ 50분 사이 임의의 값으로 정해집니다.
* 조건2: 택시기사님은 소요시간 5분 ~ 15분 사이의 승객만 매칭합니다.


**출력결과예제**

[O] 1번째 손님 (소요시간: 15분)

[X] 2번째 손님 (소요시간: 50분)

[O] 3번째 손님 (소요시간: 8분)

...

[X] 50번째 손님 (소요시간: 23분)

총 탑승 승객: 5명

```python
# Quiz 1

import random

li = [random.randint(5, 51) for i in range(50)]

n = 0

for i in range(50):
    if li[i] >= 5 and li[i] <= 15:
        print(f'[O] {i + 1}번째 손님 (소요시간: {li[i]}분)')
        n += 1
    else:
        print(f'[X] {i + 1}번째 손님 (소요시간: {li[i]}분)')
        
print(f'총 탑승 승객: {n}명')
```

#**Quiz 2**
표준 체중을 구하는 프로그램을 작성하세요.

남녀 성별에 따른 표준 체중 공식:

(남성) 키(m) x 키(m) x 22

(여성) 키(m) x 키(m) x 21

* 조건 1: 표준 체중은 별도의 함수 내에서 계산할 것

  - 함수명: std_weight

  - 전달값: 키(height), 성별(gender)


* 조건 2: 표준 체중은 소수점 둘째자리까지 표시할 것


**출력결과예제**

키 175 cm 남성의 표준 체중은 67.38kg 입니다.

```python
# Quiz 2

def std_weight(height, gender):
    if gender == 'm':value = 22
    else:value = 21
    weight = height * height * value
    return weight

setting = {'gender':'m', 'height':1.75}

print(f'키 {int(setting["height"] * 100)} cm ', end='') 
print('남성의', end='') if setting["gender"] == 'm' else print('여성의', end='')
print(f' 표준 체중은 {std_weight(**setting):0.2f}kg 입니다.')
```

#**Quiz 3**

A 회사에서는 매주 1회 작성해야 하는 보고서가 있습니다.

1주차부터 10주차까지의 보고서 양식 파일을 만드는 프로그램을 작성하세요.

보고서는 항상 아래와 같은 양식으로 출력되어야 합니다.


=== X 주차 주간보고 ===

부서 : 

이름 : 

업무 요약 : 

* 조건: 파일명은 '1주차.txt', '2주차.txt', ... 와 같이 만듭니다.

```python
# Quiz 3

for i in range(1, 11):
    f = open('./' + str(i) + '주차.txt', 'w', encoding='utf8')
    f.writelines('=== ' + str(i) +' 주차 주간보고 ===\n\n')
    f.writelines('부서:\n\n')
    f.writelines('이름:\n\n')
    f.writelines('업무 요약:')
    f.close()
```
