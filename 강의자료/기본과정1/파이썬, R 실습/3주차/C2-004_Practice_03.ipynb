#**Quiz 1**
주어진 코드를 활용하여 부동산 프로그램을 작성하세요.


**출력예제**

총 3개의 매물이 있습니다.

강남 아파트 매매 20억 (2010년 준공)

마포 오피스텔 전세 10억 (2007년 준공)

송파 빌라 월세 1000/100 (2000년 준공)

```python
# Quiz 1

# [코드]
class House:
  location = ''
  house_type = ''
  deal_type = ''
  price = 0
  completion_year = 0
  
  # 매물 초기화
  def __init__(self, location, house_type, deal_type, price, completion_year):
    self.location=location
    self.house_type=house_type
    self.deal_type=deal_type
    self.price=price
    self.completion_year=completion_year

  # 매물 정보 표시
  def show_detail(self):
    print(self.location, self.house_type, self.deal_type, self.price, self.completion_year, '\n')

content1 = {'location':'강남', 'house_type':'아파트', 'deal_type':'매매', 'price':'20억', 'completion_year':'(2010년 준공)'}
content2 = {'location':'마포', 'house_type':'오피스텔', 'deal_type':'전세', 'price':'10억', 'completion_year':'(2007년 준공)'}
content3 = {'location':'송파', 'house_type':'빌라', 'deal_type':'월세', 'price':'1000/10', 'completion_year':'(2000년 준공)'}

h = [House(**content1), House(**content2), House(**content3)]

print('총 3개의 매물이 있습니다.\n')
for i in h:
  i.show_detail()
```

#**Quiz 2**
동네에 항상 대기 손님이 있는 치킨집이 있습니다.

대기 손님의 치킨 요리 시간을 줄이고자 자동 주문 시스템을 제작하였습니다.

아래 코드를 확인하고 적절한 예외처리 구문을 넣으세요.

* 조건 1: 1보다 작거나 숫자가 아닌 입력값이 들어올 때는 ValueError 로 처리

  - 출력 메세지: "잘못된 값을 입력하였습니다."

* 조건 2: 하루에 팔 수 있는 치킨은 총 60마리로 한정

 - 치킨 소진 시 사용자 정의 에러 [SoldOutError]를 발생시키고 프로그램 종료
 - 출력 메세지: "재고가 소진되어 더 이상 주문을 받지 않습니다."

```python
# Quiz 2

# [코드]

class SoldOutError(Exception):
  pass


chicken = 60
waiting = 1 # 현재 홀은 만석. 대기번호 1부터 시작

while(True):
  try:
    print("[남은 치킨: {0}]".format(chicken))
    
    order = int(input("치킨 몇 마리 주문하시겠습니까? "))
    
    if order < 1:
      raise ValueError
    
    if order > chicken:
      print("재료가 부족합니다.")
      raise SoldOutError
      
    else:
      print("[대기번호 {0}] {1}마리 주문이 완료되었습니다.".format(waiting, order))
      waiting += 1
      chicken -= order
  except ValueError:
    print('잘못된 값을 입력하였습니다.')
  except SoldOutError:
    print('재고가 소진되어 더 이상 주문을 받지 않습니다.')
    break
```
