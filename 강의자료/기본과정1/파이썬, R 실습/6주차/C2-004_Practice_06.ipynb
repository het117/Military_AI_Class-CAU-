```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글폰트 지정
matplotlib.rcParams['font.family'] = 'NanumBarunGothic'
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.unicode_minus'] = False

font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
# 데이터 불러오기

dir = './'

df_m=pd.read_excel(dir + "202208_202208_연령별인구현황_월간.xlsx", skiprows=3, index_col='행정기관', usecols='B, E:Y')   # 남자 데이터
df_f=pd.read_excel(dir + "202208_202208_연령별인구현황_월간.xlsx", skiprows=3, index_col='행정기관', usecols='B, AB:AV') # 여자 데이터
df_m
# df_f
```

```python
# Practice

plt.figure(figsize=(10, 10))
plt.title('2022년 8월 인구 피라미드')
plt.scatter(df_m)
plt.show()
plt.savefig('2022_08_인구피라미드.png', dpi=300)
```
