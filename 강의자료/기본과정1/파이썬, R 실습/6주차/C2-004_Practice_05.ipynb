```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'NanumBarunGothic'
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.unicode_minus'] = False

from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

data = {
    '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1271], # [단위: 만 명]
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

## Quiz 1) 영화 데이터를 이용하여 x축: 영화, y축: 평점인 막대 그래프를 그리세요.

```python
# Quiz 1

plt.figure(figsize=(15, 5))
plt.bar(df['영화'], df['평점'])
plt.show
```

## Quiz 2) 앞에서 그린 막대 그래프에, 아래 제시된 세부 사항을 적용하세요.

* 제목: 국내영화 Top 8 평점
* x축 label: 영화 (90도 회전)
* y축 label: 평점

```python
# Quiz 2
plt.figure(figsize=(15, 5))
plt.bar(df['영화'], df['평점'])
plt.title('국내영화 Top 8 평점')
plt.xlabel('영화')
plt.ylabel('평점')
plt.xticks(rotation=90)
plt.show
```

## Quiz 3) 개봉 연도별 평점 변화 추이를 꺾은선 그래프로 그리세요.

```python
# Quiz 3 
so = df.sort_values('개봉연도', ascending=True)
plt.figure(figsize=(15, 5))
plt.plot(so['개봉연도'], so['평점'])
plt.xlabel('개봉연도')
plt.ylabel('평점')
plt.show
```

## Quiz 4) 앞에서 그린 꺾은선 그래프에, 아래 제시된 세부 사항을 적용하세요.

* 마커: 'o'
* x축 눈금: 5년 단위(2005, 2010, 2015, 2020)
* y축 범위: 최소 7, 최대 10

```python
# Quiz 4 
plt.figure(figsize=(15, 5))
plt.plot(so['개봉연도'], so['평점'], marker='o')
plt.xlabel('개봉연도')
plt.ylabel('평점')
plt.ylim([7, 10])
plt.xticks([2005, 2010, 2015, 2020])
plt.show
```

# Quiz 5) 평점 9점 이상인 영화의 비율을 확인할 수 있는 원 그래프를 아래 제시된 세부 사항을 적용하여 그리세요.

* label: 9점 이상, 9점 미만
* 퍼센트: 소수점 첫째자리까지 표시
* 범례(legend): 그래프 우측에 표시

```python
# Quiz 5
m = len(df)
count = len(df.loc[df['평점'] >= 9]) / m

values = [count, 1-count]
labels = ['평점 9점 이상', '평점 9점 미만']
plt.pie(values, labels=labels, autopct='%.1f%%')
plt.legend(loc=(1, 0.4))
plt.show()
```
