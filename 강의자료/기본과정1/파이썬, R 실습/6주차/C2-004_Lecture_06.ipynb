```python
# Bar 그래프 생성

fig, ax1 = plt.subplots(figsize=(12,5))
ax1.bar(df.index, df['등록대수(만대)'], color='#ff812d')

ax2 = ax1.twinx() # x축을 공유
ax2.plot(df.index, df['전년대비 증가대수(천대)'], color='#ffd100')

plt.show()
```

```python
# Label 설정

fig, ax1 = plt.subplots(figsize=(12,5))
ax1.set_ylabel('총등록대수(만대)')
ax1.bar(df.index, df['등록대수(만대)'], color='#ff812d')

ax2 = ax1.twinx() # x축을 공유
ax2.set_ylabel('년 증가대수(천대)')
ax2.plot(df.index, df['전년대비 증가대수(천대)'], color='#ffd100')

plt.show()
```

```python
# y축 범위 및 눈금 설정

fig, ax1 = plt.subplots(figsize=(12,5))
ax1.set_ylabel('총등록대수(만대)')
ax1.set_yticks([1750, 2000, 2250, 2500])
ax1.set_ylim(1750, 2500)
ax1.bar(df.index, df['등록대수(만대)'], color='#ff812d')

ax2 = ax1.twinx() # x축을 공유
ax2.set_ylabel('년 증가대수(천대)')
ax2.set_ylim(250, 1000)
ax2.set_yticks([250, 500, 750, 1000])
ax2.plot(df.index, df['전년대비 증가대수(천대)'], color='#ffd100')

plt.show()
```

```python
# 값 표기

fig, ax1 = plt.subplots(figsize=(12,5))
ax1.set_ylabel('총등록대수(만대)')
ax1.set_yticks([1750, 2000, 2250, 2500])
ax1.set_ylim(1750, 2500)
ax1.bar(df.index, df['등록대수(만대)'], color='#ff812d')

for idx1, val1 in enumerate(df['등록대수(만대)']):
  ax1.text(idx1, val1 + 12, val1, ha='center')

ax2 = ax1.twinx() # x축을 공유
ax2.set_ylabel('년 증가대수(천대)')
ax2.set_ylim(250, 1000)
ax2.set_yticks([250, 500, 750, 1000])
ax2.plot(df.index, df['전년대비 증가대수(천대)'], color='#ffd100')

plt.show()
```

```python
# 선 그래프 스타일 설정

fig, ax1 = plt.subplots(figsize=(12,5))
ax1.set_ylabel('총등록대수(만대)')
ax1.set_yticks([1750, 2000, 2250, 2500])
ax1.set_ylim(1750, 2500)
ax1.bar(df.index, df['등록대수(만대)'], color='#ff812d')

for idx1, val1 in enumerate(df['등록대수(만대)']):
  ax1.text(idx1, val1 + 12, val1, ha='center')

ax2 = ax1.twinx() # x축을 공유
ax2.set_ylabel('년 증가대수(천대)')
ax2.set_ylim(250, 1000)
ax2.set_yticks([250, 500, 750, 1000])
ax2.plot(df.index, df['전년대비 증가대수(천대)'], color='#ffd100', marker='o', ms=12, lw=5, mec='w', mew=2.5)

plt.show()
```

```python
# 그래프 제목 및 그리드 삽입

fig, ax1 = plt.subplots(figsize=(12,5))
fig.suptitle('자동차 등록대수')

ax1.set_ylabel('총등록대수(만대)')
ax1.set_yticks([1750, 2000, 2250, 2500])
ax1.set_ylim(1750, 2500)
ax1.bar(df.index, df['등록대수(만대)'], color='#ff812d')

for idx1, val1 in enumerate(df['등록대수(만대)']):
  ax1.text(idx1, val1 + 12, val1, ha='center')

ax2 = ax1.twinx() # x축을 공유
ax2.set_ylabel('년 증가대수(천대)')
ax2.set_ylim(250, 1000)
ax2.set_yticks([250, 500, 750, 1000])
ax2.plot(df.index, df['전년대비 증가대수(천대)'], color='#ffd100', marker='o', ms=12, lw=5, mec='w', mew=2.5)

plt.grid(alpha=0.3)

plt.show()
```

#**[Project] 계급별 병사봉급 추이**

```python
# https://www.index.go.kr/potal/main/EachDtlPageDetail.do?idx_cd=1700&param=013

import pandas as pd

dir = '/content/drive/MyDrive/Colab Notebooks/C2-004_Python/Lecture_06/'

df=pd.read_excel(dir + "stat_170001.xlsx", skiprows=2, nrows=4, index_col=1)
df
```

```python
df = df.T
df = df[1:11]
df = df.astype(int)
df
```

```python
# Visualization

plt.figure(figsize=(12, 5))

plt.plot(df.index, df['병장'])
plt.plot(df.index, df['상병'])
plt.plot(df.index, df['일병'])
plt.plot(df.index, df['이병'])

plt.show()
```

```python
# Label 설정

plt.figure(figsize=(12, 5))

plt.ylabel('봉급액(원)')

plt.plot(df.index, df['병장'])
plt.plot(df.index, df['상병'])
plt.plot(df.index, df['일병'])
plt.plot(df.index, df['이병'])

plt.show()
```

```python
# Y축 범위 설정

plt.figure(figsize=(12, 5))

plt.ylabel('봉급액(원)')
plt.ylim(0, 750000)
plt.yticks([0, 250000, 500000, 750000])

plt.plot(df.index, df['병장'])
plt.plot(df.index, df['상병'])
plt.plot(df.index, df['일병'])
plt.plot(df.index, df['이병'])

plt.show()
```

```python
# 그래프 스타일 적용

plt.figure(figsize=(12, 5))

plt.ylabel('봉급액(원)')
plt.ylim(0, 750000)
plt.yticks([0, 250000, 500000, 750000])

plt.plot(df.index, df['병장'], label='병장', marker='d', ms=13, mec='w', mew=2, lw=3)
plt.plot(df.index, df['상병'], label='상병', marker='o', ms=13, mec='w', mew=2, lw=3, ls=':')
plt.plot(df.index, df['일병'], label='일병', marker='^', ms=13, mec='w', mew=2, lw=3, ls='--')
plt.plot(df.index, df['이병'], label='이병', marker='+', ms=13, mew=2, lw=3, ls='-.')

plt.show()
```

```python
#  범례(legend) 적용

plt.figure(figsize=(12, 5))

plt.ylabel('봉급액(원)')
plt.ylim(0, 750000)
plt.yticks([0, 250000, 500000, 750000])

plt.plot(df.index, df['병장'], label='병장', marker='d', ms=13, mec='w', mew=2, lw=3)
plt.plot(df.index, df['상병'], label='상병', marker='o', ms=13, mec='w', mew=2, lw=3, ls=':')
plt.plot(df.index, df['일병'], label='일병', marker='^', ms=13, mec='w', mew=2, lw=3, ls='--')
plt.plot(df.index, df['이병'], label='이병', marker='+', ms=13, mew=2, lw=3, ls='-.')

plt.legend(ncol=4, loc=(0.25, -0.2))

plt.show()
```

```python
#  그래프 제목 및 그리드 선 삽입

plt.figure(figsize=(12, 5))

plt.ylabel('봉급액(원)')
plt.ylim(0, 750000)
plt.yticks([0, 250000, 500000, 750000])

plt.plot(df.index, df['병장'], label='병장', marker='d', ms=13, mec='w', mew=2, lw=3)
plt.plot(df.index, df['상병'], label='상병', marker='o', ms=13, mec='w', mew=2, lw=3, ls=':')
plt.plot(df.index, df['일병'], label='일병', marker='^', ms=13, mec='w', mew=2, lw=3, ls='--')
plt.plot(df.index, df['이병'], label='이병', marker='+', ms=13, mew=2, lw=3, ls='-.')

plt.legend(ncol=4, loc=(0.25, -0.2))

plt.title('병사봉급추이')
plt.grid(axis='y', alpha=0.3)

plt.show()
```
