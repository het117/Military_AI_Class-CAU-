## 2-2. Subword Tokenizer

* **Vocabulary(단어 집합):** 서로 다른 단어들의 집합

* **UNK(Unknown Token):** 단어 집합에 없는 단어

* **OOV(Out-Of-Vocabulary):** 단어 집합에 존재하지 않는 단어가 발생

* **Subword Segmentation:** 하나의 단어를 더 작은 단위의 서브 워드로 분리(ex. birthplace = birth + place) → OOV나 희귀 단어, 신조어 등의 문제를 완화할 수 있음.


### **# BPE(Byte Pair Encoding)**

* 1994년에 제안된 데이터 압축 알고리즘

* 이후 Subword Segmentation Algorithm으로 응용

* 기본적으로, 연속적으로 가장 많이 등장한 글자 쌍을 찾아 하나의 글자로 병합하는 방식 수행



**[예제]**

**aaabdaaabac**


1)

가장 자주 등장하는 byte pair는 'aa' 

→'aa'라는 byte pair를 하나의 byte인 'Z'로 치환

**ZabdZabac** (Z=aa)




2)

가장 자주 등장하는 byte pair는 'ab' ('Za'도 있지만, 치환되지 않은 byte에 우선순위를 부여하는 것으로 가정)

→'ab'라는 byte pair를 하나의 byte인 'Y'로 치환

**ZYdZYac** (Z=aa, Y=ab)




3)

가장 자주 등장하는 byte pair는 'ZY' 

→'ZY'라는 byte pair를 하나의 byte인 'X'로 치환

**XdXac** (Z=aa, Y=ab, X=ZY)




4)

더 이상 병합할 byte pair가 없으므로, 위의 결과를 최종 결과로 하여 BPE 종료

### **# 자연어 처리에서의 BPE(Byte Pair Encoding)**

* 자연어 처리에서의 BPE는 Subword Segmentation Algorithm임.

* 글자(character) 단위에서 점차적으로 단어 집합(vocabulary)를 만들어 내는 Bottom up 방식의 접근을 사용

**[예제]**

low, lower, newest, widest

**1) 기존의 접근(빈도 기반의 카운트)**

```python
'''
# dictioanry (훈련 데이터에 있는 단어와 등장 빈도수)
  → low : 5, lower : 2, newest : 6, widest : 3

# vocabulary
  → low, lower, newest, widest
'''
```

```python
# 이때 테스트 과정에서 'lowest'라는 단어가 등장한다면?
#'lowest'를 학습한 적이 없으므로, 이에 대응하지 못하는 OOV 문제 발생 !
```

**2) BPE 알고리즘**

```python
'''
# dictionary (모든 단어들을 character 단위로 분리)
  → l o w : 5,  l o w e r : 2,  n e w e s t : 6,  w i d e s t : 3

# vocabulary (초기 구성: 글자 단위로 분리된 상태)
  → l, o, w, e, r, n, s, t, i, d
'''
```

BPE는 사용자가 정한 단어 집합의 크기를 만족할 때까지 알고리즘의 동작을 반복함.
(가장 빈도수가 높은 유니그램의 쌍을 하나의 유니그램으로 통합하는 과정을 반복)

1회 - 딕셔너리를 참고하여, 빈도수가 9로 가장 높은 (e, s)의 쌍을 es로 통합

```python
# (e, s), (s, t) -> 둘다 빈도수가 9
# 실제 BPE에서는 병합 우선순위가 존재.
# 여기서는 먼저 나온 pair에 우선순위 부여하는 것으로 설정
```

```python
'''
# dictionary update!
l o w : 5,
l o w e r : 2,
n e w es t : 6,
w i d es t : 3

# vocabulary update!
l, o, w, e, r, n, s, t, i, d, es
'''
```

2회 - 빈도수가 9로 가장 높은 (es, t)의 쌍을 est로 통합

```python
'''
# dictionary update!
l o w : 5,
l o w e r : 2,
n e w est : 6,
w i d est : 3

# vocabulary update!
l, o, w, e, r, n, s, t, i, d, es, est
'''
```

3회 - 빈도수가 7로 가장 높은 (l, o)의 쌍을 lo로 통합

```python
'''
# dictionary update!
lo w : 5,
lo w e r : 2,
n e w est : 6,
w i d est : 3

# vocabulary update!
l, o, w, e, r, n, s, t, i, d, es, est, lo
'''
```

10회 반복했을 때

```python
'''
# dictionary update!
low : 5,
low e r : 2,
newest : 6,
widest : 3

# vocabulary update!
l, o, w, e, r, n, s, t, i, d, es, est, lo, low, ne, new, newest, wi, wid, widest
'''
```

**코드실습**

```python
dictionary = {'l o w' : 5,
         'l o w e r' : 2,
         'n e w e s t':6,
         'w i d e s t':3
         }
```

```python
import re, collections

def get_stats(dictionary):    # 유니그램의 pair들의 빈도수를 카운트
    pairs = collections.defaultdict(int)
    for word, freq in dictionary.items():
        symbols = word.split()
        for i in range(len(symbols)-1):
            pairs[symbols[i],symbols[i+1]] += freq
    print('현재 pair들의 빈도수 :', pairs)
    return pairs


def merge_dictionary(pair, v_in):    # dictionary update _ 'n e w e s t' → 'n e w es t'
    v_out = {}
    bigram = re.escape(' '.join(pair))    # if pair=('l','o'), bigram=l\ o
    p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
    for word in v_in:
        w_out = p.sub(''.join(pair), word)
        v_out[w_out] = v_in[word]
    return v_out
```

```python
from IPython.display import display, Markdown, Latex

num_merges = 10    # BPE 수행 횟수

bpe_codes = {}
bpe_codes_reverse = {}

for i in range(num_merges):
    display(Markdown("### Iteration {}".format(i + 1)))
    pairs = get_stats(dictionary)
    best = max(pairs, key=pairs.get)
    dictionary = merge_dictionary(best, dictionary)

    bpe_codes[best] = i
    bpe_codes_reverse[best[0] + best[1]] = best

    print("new merge: {}".format(best))
    print("dictionary: {}".format(dictionary))
```

## **Assignment(week1)

**[문제] 'low', 'lower', 'newest', 'widest'에 대하여 BPE를 사용해서 단어집합을 구성하였다. 이때 새로 등장한 단어 'lowest'에 대하여 토큰화를 수행해보자.**

**[Hint]**

1.   lowest를 글자 단위로 분해 → l, o, w, e, s, t
2.   앞서 업데이트된 단어집합을 사용해서 pair를 통합
(이때, 통합을 위한 우선순위는 bpe_codes를 사용(value값이 작을수록 대응하는 key의 우선순위가 높음))

```python
def get_pairs(word):
    """Return set of symbol pairs in a word.
    Word is represented as a tuple of symbols (symbols being variable-length strings).
    """
    pairs = set()
    prev_char = word[0]
    for char in word[1:]:
        pairs.add((prev_char, char))
        prev_char = char
    return pairs

def encode(origal):
    """Encode word based on list of BPE merge operations, which are applied consecutively"""

    tokenized_word = tuple(origal) + ('</w>',)
        
    display(Markdown("__word split into characters:__ <tt>{}</tt>".format(tokenized_word)))

    pairs = get_pairs(tokenized_word)
    print(pairs)

    if not pairs:
        return origal

    iteration = 0
    while True:
        iteration += 1
        display(Markdown("__Iteration {}:__".format(iteration)))

        print("bigrams in the word: {}".format(pairs))
        bigram = min(pairs, key = lambda pair: bpe_codes.get(pair, float('inf')))
        print("candidate for merging: {}".format(bigram))
        if bigram not in bpe_codes:
            display(Markdown("__Candidate not in BPE merges, algorithm stops.__"))
            break
        first, second = bigram
        new_word = []
        i = 0
        while i < len(tokenized_word):
            try:
                j = tokenized_word.index(first, i)
                new_word.extend(tokenized_word[i:j])
                i = j
            except:
                new_word.extend(tokenized_word[i:])
                break

            if tokenized_word[i] == first and i < len(tokenized_word)-1 and tokenized_word[i+1] == second:
                new_word.append(first+second)
                i += 2
            else:
                new_word.append(tokenized_word[i])
                i += 1
        new_word = tuple(new_word)
        tokenized_word = new_word
        print("word after merging: {}".format(tokenized_word))
        if len(tokenized_word) == 1:
            break
        else:
            pairs = get_pairs(tokenized_word)

    # don't print end-of-word symbols
    if tokenized_word[-1] == '</w>':
        tokenized_word = tokenized_word[:-1]
    elif tokenized_word[-1].endswith('</w>'):
        tokenized_word = tokenized_word[:-1] + (tokenized_word[-1].replace('</w>',''),)

    return tokenized_word
```
