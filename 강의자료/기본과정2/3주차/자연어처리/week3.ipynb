**[과제] Neural Network로 XOR 게이트 구현하기**

**[XOR 게이트]**

입력값 두 개가 서로 다른 값을 갖고 있을때에만 출력값이 1이 되고, 입력값 두 개가 서로 같은 값을 가지면 출력값이 0이 되는 게이트

```python
import torch
import torch.nn as nn
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to(device) 

Y = torch.FloatTensor([[0], [1], [1], [0]]).to(device)
```

```python
# and gate
def and_gate(x1, x2):
    w1, w2 = 0.5, 0.5
    b=-0.7

    result = x1*w1 + x2*w2 + b
    
    return 0 if result <= 0 else 1

# or gate
def or_gate(x1, x2):
    w1, w2 = 0.7, 0.3
    b=-0.1

    result = x1*w1 + x2*w2 + b
    
    return 0 if result <= 0 else 1

# nand gate
def nand_gate(x1, x2):
    w1, w2 = -0.5, -0.5
    b=0.7

    result = x1*w1 + x2*w2 + b
    
    return 0 if result <= 0 else 1

# XOR gate
def XOR(x1, x2):
    s1 = nand_gate(x1, x2)
    s2 = or_gate(x1, x2)
    result = and_gate(s1, s2)
    return result

for x1, x2 in X:
    nand_result = XOR(x1, x2)
    print(f"입력값: {x1} {x2} / 결과: {nand_result}")
```

```python
class XOR(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = torch.nn.Linear(2, 5)
        self.layer2 = torch.nn.Linear(5, 1)

    def forward(self, x):
        x = torch.relu(self.layer1(x))
        x = torch.sigmoid(self.layer2(x))
        return x


model = XOR().to(device)

criterion = torch.nn.BCELoss().to(device)
optimizer = torch.optim.SGD(model.parameters(), lr=0.02)

for step in range(10000): 
    optimizer.zero_grad()
    output = model(X).to(device)

    loss = criterion(output, Y)
    loss.backward()
    optimizer.step()
    if step % 5==0:
        print(step, loss.item())

# 예측값
predicted = (output>0.5).float()
accuracy = (predicted==Y).float().mean()
print('모델의 출력값(Hypothesis): ', output.detach().cpu().numpy())
print('모델의 예측값(Predicted): ', predicted.detach().cpu().numpy())
print('실제값(Y): ', Y.cpu().numpy())
print('정확도(Accuracy): ', accuracy.item())
```
